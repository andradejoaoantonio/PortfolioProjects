# BMI Calculator

Formulas: https://www.thecalculatorsite.com/articles/health/bmi-formula-for-bmi-calculations.php
BMI information: https://my.clevelandclinic.org/health/articles/9464-body-mass-index-bmi

```python
# Body Mass Index (BMI) (Imperial) = (weight * 703) / (height * height)
# Body Mass Index (BMI) (Metric) = weight / (height * height)
```

```python
# BMI (Imperial)
name = input("enter your name: ")

weight = int(input("enter you weight in pounds: "))

height = int(input("enter your height in inches: "))

BMI = (weight * 703) / (height * height)

print(BMI)

if BMI > 0:
    if(BMI < 18.5):
        print(name +", you are underweight.")
    elif (BMI <= 24.9):
        print(name +", you are normal weight.")
    elif (BMI < 29.9):
        print(name +", you are overweight.")
    elif (BMI < 34.9):
        print(name +", you are obese.")
    elif (BMI < 39.9):
        print(name +", you are severely obese.")
    else:
        print(name +", you are morbidly obese.")
else:
    print("Enter valid input")
```

```python
# BMI (Metric)
name = input("enter your name: ")

weight = int(input("enter you weight (kgs): "))

height = float(input("enter your height (m): "))

BMI = weight / (height*height)

print(BMI)

if BMI > 0:
    if(BMI < 18.5):
        print(name +", you are underweight.")
    elif (BMI <= 24.9):
        print(name +", you are in the optimum range.")
    elif (BMI < 29.9):
        print(name +", you are overweight.")
    elif (BMI < 34.9):
        print(name +", you are in the class I obesity.")
    elif (BMI < 39.9):
        print(name +", you are in the class II obesity.")
    else:
        print(name +", you are in the in the class I obesity.")
else:
    print("Enter valid input")
```

```python
#Note:
Under 18.5	Underweight	Minimal
18.5 - 24.9	Optimum Range	Minimal
25 - 29.9	Overweight	Increased
30 - 34.9	Class I obesity	High
35 - 39.9	Class II obesity	Very High
40 and over	Class III obesity	Extremely High
```
