# Automatic File Sorter in Explorer

```python
# Image we have a folder with many files, different types and formats
# We want automatically place each file type into different folders

# shutil = "shell utilities" because it allows you to perform various file-related tasks
# that are similar to the commands that you might use in a Unix shell, such as copying, moving,
# renaming, and deleting files and directories.
```

```python
import os, shutil
```

```python
# r stands for "raw text"
path = r"C:/Users/Joao Andrade/Desktop/P2 Automatic File Sorter example/"
```

```python
file_name = os.listdir(path)
```

```python
os.path.exists(path + 'xlsx')
```

```python
folder_names = ['xlsx','image','text','pdf']

for loop in range(0,4):
    if not os.path.exists(path + folder_names[loop]):
        #print((path + folder_names[loop]))
        os.makedirs(path + folder_names[loop])

for file in file_name:
    if ".xlsx" in file and not os.path.exists(path + "xlsx/"+ file):
        shutil.move(path + file, path + "xlsx/" + file)
    elif ".jpg" in file and not os.path.exists(path + "image/"+ file):
        shutil.move(path + file, path + "image/" + file)
    elif ".txt" in file and not os.path.exists(path + "text/"+ file):
        shutil.move(path + file, path + "text/" + file)
    elif ".pdf" in file and not os.path.exists(path + "pdf/"+ file):
        shutil.move(path + file, path + "pdf/" + file)
```
